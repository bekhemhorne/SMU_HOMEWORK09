```python
# Clustering/Dimensionality Reduction
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read in and drop in column for myopic data specifically 
df = pd.read_csv("myopia.csv")
target = df.MYOPIC
df.drop(columns=["MYOPIC"], inplace=True)
df.head()
```

```python
#Make sure it is numeric and non null
df.info()
```

```python
#Start by checking for obvious clustering.
#Projecting the rough idea if we can make clusters
#Looks like top left and mid right will cluster

X = TSNE(learning_rate="auto").fit_transform(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:, 0], X[:, 1])
plt.title("TSNE Plot")
plt.show()
```

```python
#Initialize PCA Model
#We can see elbows at 1, 2, 3 and slightly 5.
#looks like 3 components might be a majority of variance 
pca = PCA()
pca.fit(df)

plt.plot(range(len(df.columns)), pca.explained_variance_ratio_)
plt.xlabel("# of Principal Components")
plt.ylabel("Explained Variance")
plt.title("PCA Elbow")
plt.show()
```

```python
#component 0 is 72% and so on, there is some corrilation 
list(zip(range(len(df.columns)), pca.explained_variance_ratio_))
```

```python
df.corr()
```

```python
#Do three principal components
#
pca = PCA(n_components=3)
pca.fit(df)

components = pd.DataFrame(pca.transform(df), columns=["PCA1", "PCA2", "PCA3"])

components.head()
```

```python
#Check for obivous clusters part 2.
#not very different
X = TSNE(learning_rate="auto").fit_transform(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:, 0], X[:, 1])
plt.title("TSNE Plot")
plt.show()
```

```python
#Kmeans
#We can see elbows just after 2.5 and 5.0
#not a strong elbow though
inertia = []
k_range = list(range(2, 20))

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    
    inertia.append(kmeans.inertia_)
    
#Make the plot
plt.figure(figsize=(12,8))

plt.plot(k_range, inertia)
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Plot")

plt.show()
```

```python
#KMeans cluster of 8
#A lot of overlap
#8 seems to be overfitting
km = KMeans(n_clusters=8, random_state=42)
kmeans.fit(df)
preds = kmeans.predict(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:, 0], X[:, 1], c=preds)
plt.title("TSNE Plot")
plt.show()
```

```python
#KMeans cluster of 5
#Purple and Green have interesting boundary points
km = KMeans(n_clusters=5, random_state=42)
kmeans.fit(df)
preds = kmeans.predict(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:, 0], X[:, 1], c=preds)
plt.title("TSNE Plot")
plt.show()
```

```python
#Dendrogram plot
#Green is a clear cluster
#Orange is spread out into about four or five groups, two of which could be combined to one.
mergings = linkage(df, method = "ward")

plt.figure(figsize=(20, 8))

dendrogram(mergings, 
           leaf_rotation=90,
           leaf_font_size=5)

plt.show()
```

```python
cluster = AgglomerativeClustering(
    n_clusters=5,
    affinity="euclidean",
    linkage="ward"
)

preds = cluster.fit_predict(df)
plt.figure(figsize=(12,8))

plt.scatter(X[:, 0], X[:, 1], c=preds)
plt.title("TSNE Plot")
plt.show()
```
